```python
%matplotlib inline
from collections import defaultdict
import json

import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
import pandas as pd

from matplotlib import rcParams
import matplotlib.cm as cm
import matplotlib as mpl

#colorbrewer2 Dark2 qualitative color table
dark2_colors = [(0.10588235294117647, 0.6196078431372549, 0.4666666666666667),
                (0.8509803921568627, 0.37254901960784315, 0.00784313725490196),
                (0.4588235294117647, 0.4392156862745098, 0.7019607843137254),
                (0.9058823529411765, 0.1607843137254902, 0.5411764705882353),
                (0.4, 0.6509803921568628, 0.11764705882352941),
                (0.9019607843137255, 0.6705882352941176, 0.00784313725490196),
                (0.6509803921568628, 0.4627450980392157, 0.11372549019607843)]

rcParams['figure.figsize'] = (10, 6)
rcParams['figure.dpi'] = 150
rcParams['axes.color_cycle'] = dark2_colors
rcParams['lines.linewidth'] = 2
rcParams['axes.facecolor'] = 'white'
rcParams['font.size'] = 14
rcParams['patch.edgecolor'] = 'white'
rcParams['patch.facecolor'] = dark2_colors[0]
rcParams['font.family'] = 'StixGeneral'


def remove_border(axes=None, top=False, right=False, left=True, bottom=True):
    """
    Minimize chartjunk by stripping out unnecesasry plot borders and axis ticks
    
    The top/right/left/bottom keywords toggle whether the corresponding plot border is drawn
    """
    ax = axes or plt.gca()
    ax.spines['top'].set_visible(top)
    ax.spines['right'].set_visible(right)
    ax.spines['left'].set_visible(left)
    ax.spines['bottom'].set_visible(bottom)
    
    #turn off all ticks
    ax.yaxis.set_ticks_position('none')
    ax.xaxis.set_ticks_position('none')
    
    #now re-enable visibles
    if top:
        ax.xaxis.tick_top()
    if bottom:
        ax.xaxis.tick_bottom()
    if left:
        ax.yaxis.tick_left()
    if right:
        ax.yaxis.tick_right()
        
pd.set_option('display.width', 500)
pd.set_option('display.max_columns', 100)

from IPython.display import Image as Im
from IPython.display import display
```

```python
# getting data from text file and assign to array
data=np.array([[float(j) for j in e.strip().split()] for e in open("chall.txt")])
#display data
data
```

```python
temps, pfail = data[:,0], data[:,1]    # split data into 2 separate variables for further use
plt.scatter(temps, pfail, marker = '*', color = 'blue')     # plotting scatter graph
plt.title('Temperature and o-ring failure')
axes=plt.gca()     # gca = get current axis
plt.xlabel('Temperature')
axes.grid(False)
remove_border(axes)
```

```python
from sklearn.linear_model import LogisticRegression   # binary prediction 
reg=1000.    # Inverse of Regularization strength (solving overfitting), smaller the stronger, we dont need it for only 2 factors here
clf4 = LogisticRegression(C=reg)
clf4.fit(temps.reshape(-1,1), pfail)   # fitting data to regression (create a model)
```

```python
tempsnew=np.linspace(20., 90., 15)    # generate array of 15 temperature data from 20 to 90
probs = clf4.predict_proba(tempsnew.reshape(-1,1))[:, 1]    # predict the probability for the tempsnew (based on model)
predicts = clf4.predict(tempsnew.reshape(-1,1))   # assign fail value from prediction
```

```python
plt.scatter(temps, pfail, marker = '*', color= 'blue')     # plot actual fact data first
axes=plt.gca()              # cosmetic
axes.grid(False)            # cosmetic
remove_border(axes)         # cosmetic
plt.xlabel('Temperature')
plt.ylabel('O-Ring Damage 0=No, 1=Yes')
plt.plot(tempsnew, probs, marker='o')      # plot the probability curve, marker = circle
plt.scatter(tempsnew, predicts, marker='D', color="green")   # plot where our new temperature data lie, marker = diamond
```

```python
# crosstab to compare prediction and actual result
pd.crosstab(pfail, clf4.predict(temps.reshape(-1,1)), rownames=["Actual"], colnames=["Predicted"])
```
